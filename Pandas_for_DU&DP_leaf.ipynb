#Concat Data

```python
#Concat Series
ser1 = pd.Series(['A','B','C'], index=[1,2,3])
ser2 = pd.Series(['D','E','F'], index=[1,2,3])
print(ser1)
print(ser2)
ser3 = pd.concat([ser1,ser2],ignore_index=True) #Fales ไม่ได้สร้าง index ให้ใหม่ Ture สร้าง index ให้ใหม่
ser3
```

```python
#Concat DataFrame By Rowa
df1 = pd.DataFrame({'A':['A1','A2'],
                    'B':['B1','B2']},[0,1])
df2 = pd.DataFrame({'A':['A3','A4'],
                    'B':['B3','B4']},[0,1])
```

```python
print(df1)
print(df2)
```

```python
pd.concat([df1,df2],ignore_index=False)
```

```python
pd.concat([df1,df3],axis=0,ignore_index=False,sort=False)
```

```python
pd.concat([df1,df3],axis=1)
```

#Joining Data

```python
#Joining Data
df1 = pd.DataFrame({'employee':['Bob','Jack','Lisa','Sue'],
                    'group':['Account','Engineering','Engineering','HR']})

df2 = pd.DataFrame({'name':['Lisa','Bob','Jack','Sue'],
                    'hire_date':[2004,2008,2012,2014]})

print(df1)
print(df2)
```

```python
#one to one join
df3 = pd.merge(df1,df2,how='inner',left_on='employee',right_on='name')
df3
```

#GroupBy

```python
df_groupby = df.groupby('key').mean()
df_groupby
```
